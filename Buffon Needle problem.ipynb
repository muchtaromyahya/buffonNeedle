```python
import matplotlib.pyplot as plt
import numpy as np
import random as rand

%matplotlib inline

#masukkan banyak jarum yang ingin dijatuhkan:
trials = int(input("banyak jarum yang mau dijatuhkan ? antara 0-1000 : "))

while trials < 1 or trials > 1000:
    
    trials = int(input("angka lebih dari 0 kurang dari 1000 : "))

length = float(input("panjang jarumnya ? : "))

while length > 1 or length <= 0:
    
    length = float(input("pilih angka lebih dari 0 kurang dari l: "))

fig, ax = plt.subplots(figsize=(10, 10))
ax.grid(axis='x', color = "#2653FF")

color = ""
crossed = 0
not_crossed = 0

#plot jarum dan generate jarum:
for n in range(0, trials):
    
    #Gmembuat titik tengah jarum:
    center_x = rand.uniform(0,5)
    center_y = rand.uniform(0,5)
    angle = np.radians(rand.uniform(0,180))
    
    #Calculating the endpoints of the needle:
    x1 = center_x - (np.sin(angle)*.5*length)
    x2 = center_x + (np.sin(angle)*.5*length)
    
    y1 = center_y - (np.cos(angle)*.5*length)
    y2 = center_y + (np.cos(angle)*.5*length)
    
    horiz = np.linspace(x1, x2)
    vert = np.linspace(y1, y2)
    
    #Check jika jarum menyentuh garis:
    if x1 < 0 < x2 or x1 > 0 > x2 or x1 < 1 < x2 or x1 > 1 > x2 or x1 < 2 < x2 or x1 > 2 > x2 or x1 < 3 < x2 or x1 > 3 > x2 \
    or x1 < 4 < x2 or x1 > 4 > x2 or x1 < 5 < x2 or x1 > 5 > x2:
        
        color = "#FF0000"
        crossed += 1
        
    else:
        
        color = "#008000"
        not_crossed += 1
    
    #plot jarumnya
    plt.plot(horiz,vert,color = color)

plt.xlim(-1,5.5)
plt.ylim(-1,5.5)
plt.show()

print(f"{crossed} jarum menyentuh garis, dan {not_crossed} tidak.\n")

chance = crossed/trials
calc_chance = 2*length/np.pi
calc_pi = 2*length/chance
difference = abs(np.pi - calc_pi)

print(f"{round(chance*100, 3)}% jarum menyentuh garis. yang diprediksi berdasarkan formula {round(calc_chance*100, 3)}%.")
print()
print(f"estimasi pi yang didapatkan: {calc_pi}")
print(f"estimasi perbedaan pi : {round(difference, 3)}.")
```
