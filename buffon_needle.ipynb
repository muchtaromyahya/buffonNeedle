```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initialization
n_total = np.arange(100, 10100, 100)
inner_iter = 10
d = 1       # line width
l = 0.8     # needle length
```

```python
# iteration
pi_list = []
for n in n_total:
    for j in range(inner_iter):
        tmp = []
        n_hits = 0
        for i in range(n):
            theta = np.random.rand() * (np.pi / 2)
            x = np.random.rand() * d
            if x <= (l * np.cos(theta)):
                n_hits += 1
        pi = ((2*l)/d) * (n/n_hits)
        if n == n_total[-1] and j == inner_iter-1:
            plt.scatter(n, pi, c='blue', label='$\pi_{approx.}$')
        else:
            plt.scatter(n, pi, c='blue')
        tmp.append(pi)
    pi_list.append(np.average(tmp))
pi_list
plt.plot(n_total, pi_list, c='red', linewidth=3.0, label='$\pi_{avg.}$')
plt.axhline(y=np.pi, c='black', label='$\pi_{exact}$')
plt.xlabel('N$_{total}$')
plt.ylabel('$\pi$')
plt.legend()
plt.show()
```
